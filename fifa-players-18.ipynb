## Visualizations 
- [Creating subplots](https://matplotlib.org/stable/gallery/subplots_axes_and_figures/subplots_demo.html)
- [Creating a correlation heatmap](https://medium.com/@szabo.bibor/how-to-create-a-seaborn-correlation-heatmap-in-python-834c0686b88e)
- Create a scatter plot for something correlated

```python
corr = df.corr() # to see corr with a specific colummn # df.corr()['col_name']
```

```python
# create a correlation heatmap
fig, ax = plt.subplots(figsize=[15, 10])
mask = np.triu(np.ones_like(corr, dtype=bool))
sns.heatmap(corr, mask=mask)
plt.show()
```

```python
fig, (ax1, ax2) = plt.subplots(2, figsize=[15, 10]) # define your axis

sns.scatterplot(x = df['Dribbling'], y = df['Positioning'], hue=df['Age'], ax=ax1)
ax1.set_title('Dribbling vs. Positioning')

sns.scatterplot(x = df['Dribbling'], y = df['GKDiving'], hue=df['Age'], ax=ax2)
plt.title('Dribbling vs. GKDiving')

fig.suptitle('Dribblings vs. Positioning and GK Diving by Age') # give your entire figure a title 
plt.show()
```
