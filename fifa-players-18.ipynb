## Visualizations 
- [Creating subplots](https://matplotlib.org/stable/gallery/subplots_axes_and_figures/subplots_demo.html)
- [Creating a correlation heatmap](https://medium.com/@szabo.bibor/how-to-create-a-seaborn-correlation-heatmap-in-python-834c0686b88e)

```python
corr = df.corr() # to see corr with a specific colummn # df.corr()['col_name']
```

```python
# create a correlation heatmap
fig, ax = plt.subplots(figsize=[15, 10])
mask = np.triu(np.ones_like(corr, dtype=bool))
sns.heatmap(corr, mask=mask)
plt.show()
```

```python
fig, ax = plt.subplots(figsize=[15, 15])
sns.scatterplot(x = df['Age'], y = df['International Reputation'])
plt.show()
```
