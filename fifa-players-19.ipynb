```python
df['Wage']
```

```python
# what type of money are they making? 
money = []
for w in df['Wage']:
    if (w[-1] not in money):
        money.append(w[-1])
#     print(w[-1])
money
```

```python
# is there a better way to do this? 
wages = []
for w in df['Wage']:
#     wage = w.strip('€')
    wage = int(w.strip('€').strip('K')) * 1000
    wages.append(wage)
```

```python
df['Wages_Int'] = wages
```

```python
# normalize
scaler = StandardScaler()
scaled_wages = scaler.fit_transform(df['Wages_Int'].to_numpy().reshape(-1, 1))
```

```python
fig = plt.subplots(figsize=[15, 10]) # define your axis
sns.histplot(df['Penalties'])
plt.show()
```

```python
corr = df.corr() # to see corr with a specific colummn # df.corr()['col_name']
```

```python
# create a correlation heatmap
cmap = sns.diverging_palette(220, 20, as_cmap=True)
fig, ax = plt.subplots(figsize=[15, 10])
mask = np.triu(np.ones_like(corr, dtype=bool))
sns.heatmap(corr, mask=mask, cmap=cmap)
plt.show()
```
